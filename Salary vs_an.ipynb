```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

```python
nba_data = pd.read_csv("../Resources/cleaned_df.csv")
```

```python
nba_data.head()
```

```python
x_PTS = nba_data['PTS']
y_Salary = nba_data['SALARY']
plt.scatter(x_PTS,y_Salary,marker = 'o',edgecolors = 'blue')
plt.title("Points vs. Salary")
plt.xlabel("Points")
plt.ylabel("Salary")
plt.grid()
plt.yticks(np.arange(0,50000000,10000000))
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_PTS, y_Salary)
regress_values = x_PTS * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(x_PTS,regress_values,"r-")
plt.annotate(line_eq,(-50,10),fontsize=15,color="red")
print(f"The r-squared is: {rvalue}")
```
